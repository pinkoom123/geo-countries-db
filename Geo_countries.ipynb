```python

"""
Country Codes Dataset
Comprehensive global country data including ISO codes, regions, capitals, currencies, GDP, and bounding boxes.

Author: Patrick Kobina Inkoom Feb,2026
"""


import json

# Read JSON dictionary {code: {data...}}
with open('country_codes', 'r') as f:
    country_dict = json.load(f)

countries = list(country_dict.values())

def get_country_by_alpha2(code: str):
    """Get by ISO alpha-2."""
    code = code.upper()
    return country_dict.get(code.lower())

def get_country_by_name(name: str):
    """Find by name."""
    name = name.lower()
    for country in countries:
        if name in country.get('name', '').lower():
            return country
    return None

def print_country_info(country):
    """Print full country details."""
    if not country:
        print("Country not found.")
        return
    print(f"{country['name']} | Capital: {country['capital']} | "
          f"Area: {country.get('area-km2', 'N/A')} km² | Sub_region: {country.get('sub-region', 'N/A')} | "
          f"Bounding_boxes: {country.get('boundingBox', 'N/A')}")

print(f" Loaded {len(countries)} countries.")
print("\nUsage examples:")
print_country_info(get_country_by_alpha2('GH'))  # Ghana
print_country_info(get_country_by_alpha2('US'))  # USA  
print_country_info(get_country_by_name('Nigeria'))  # Nigeria
```
